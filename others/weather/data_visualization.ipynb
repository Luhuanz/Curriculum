```python

import pandas
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
# 
# plt.style.use('ggplot') 

# 
data1 = pd.read_csv('zhenjiang.csv')

# 
# 
plt.style.use('ggplot')
plt.rcParams['font.sans-serif'] = ['SimHei']
# 
plt.rcParams['axes.unicode_minus'] = False
# 
datalsit = pandas.read_csv('zhenjiang.csv', encoding='utf-8')

# 
# split/
datalsit[''] = datalsit['/'].str.split('/', expand=True)[0]
datalsit[''] = datalsit['/'].str.split('/', expand=True)[1]
# 
datalsit[''] = datalsit[''].map(lambda x: str(x.replace('', '')))
datalsit[''] = datalsit[''].map(lambda x: str(x.replace('', '')))

dates = datalsit['']
highs = datalsit['']
lows = datalsit['']

# 
# 
fig = plt.figure(dpi=128,figsize=(10,6))
# 
plt.plot(dates,highs,c='red',alpha=0.5)
plt.plot(dates,lows,c='blue',alpha=0.5)
# 
plt.fill_between(dates,highs,lows,facecolor='blue',alpha=0.2)

# 
# 
plt.title('2023',fontsize=24)
plt.xlabel('',fontsize=12)
# x  30 rotation=30
fig.autofmt_xdate()
plt.ylabel('',fontsize=12)
# 
plt.tick_params(axis='both',which='major',labelsize=10)
#  101
plt.xticks(dates[::10])

plt.show()
```

```python
from pyecharts.charts import Pie, Line, Radar #     
from pyecharts import options as opts   # 
from pyecharts.globals import ThemeType # 
from pyecharts.charts import Page # html
from collections import Counter
```

```python
df['2']=df[''].apply(lambda x: x.strftime('%Y-%m-%d')).tolist() #xtimestamp
```

```python
df['2']
```

```python
# ''
df[['', '']] = df['/'].str.split('/', expand=True)
df[''] = df[''].str.replace('', '').astype(int)
df[''] = df[''].str.replace('', '').astype(int)
# 1. 
line_chart = Line(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))#    https://pyecharts.org/
#  strftime 
date_strings = df[''].dt.strftime('%m-%d').tolist()
line_chart.add_xaxis(date_strings)
line_chart.add_yaxis("", df[''].tolist(), is_smooth=True)
line_chart.add_yaxis("", df[''].tolist(), is_smooth=True)
# 
line_chart.set_global_opts(title_opts=opts.TitleOpts(title="2023"),
                     xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=45)),# x45^o
                     yaxis_opts=opts.AxisOpts(name=""),
                     legend_opts=opts.LegendOpts(is_show=True))

# 
line_chart.set_series_opts(label_opts=opts.LabelOpts(is_show=False),linestyle_opts=opts.LineStyleOpts(width=2))
# 2. - 
weather_conditions = df[''].value_counts().to_dict() #  
pie_chart = Pie()
pie_chart.add("", [list(z) for z in weather_conditions.items()])
pie_chart.set_global_opts(
    title_opts=opts.TitleOpts(title=""),
    legend_opts=opts.LegendOpts(is_show=False) # 
)
pie_chart.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))

# 3. - 
wind_directions = df['(/)'].str.split('/', expand=True).stack()
wind_counts = Counter(wind_directions)

# 
radar_data = [list(wind_counts.values())]
radar_schema = [{"name": wind, "max": max(wind_counts.values())} for wind in wind_counts.keys()]

# 
radar_chart = Radar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
radar_chart.add_schema(schema=radar_schema)
radar_chart.add("", radar_data, linestyle_opts=opts.LineStyleOpts(color="#f9713c", width=1))

# 
radar_chart.set_global_opts(title_opts=opts.TitleOpts(title=""))
```

```python
#  Page 
page = Page()

# 
page.add(
    line_chart,
    pie_chart,
    radar_chart
)
#  HTML 
page.render("charts.html")
```
