```python
# 定义分词函数
# 添加不希望被分割的词语
jieba.add_word('超能力')
def segment_text(text):
    words = jieba.lcut(text)
    return ' '.join(words)
```

```python
temp=df['弱点'].apply(segment_text)
```

```python
temp=temp.str.split(expand=True) 
```

```python
df['弱点']=temp[0]
```
